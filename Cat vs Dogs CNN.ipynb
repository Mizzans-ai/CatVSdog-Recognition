```python
import numpy as np
import keras
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_path =r"C:\Users\fahad\Desktop\catvsdog train"
test_path = r"C:\Users\fahad\Desktop\test catvsdog"
```

```python
train_data = ImageDataGenerator(rescale = 1./255)
test_data = ImageDataGenerator(rescale = 1./255)
```

```python
X = train_data.flow_from_directory(train_path,target_size =(150,150),batch_size=35,class_mode = 'binary')
y = test_data.flow_from_directory(test_path,target_size =(150,150),batch_size=35,class_mode = 'binary')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D,Flatten,Dense
```

```python
model = Sequential()
```

```python
model.add(Conv2D(filters = 32 , kernel_size = (4,4),input_shape = (150,150,3),activation = 'relu'))
model.add(MaxPool2D(pool_size = (2,2)))
model.add(Flatten())
```

```python
model.add(Dense(units = 200,activation ='relu'))
model.add(Dense(units = 1,activation ='sigmoid'))
```

```python
model.compile(optimizer = "adam",loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(X,validation_data = y, epochs = 1)
```

```python
model.save('catsvsdogs.keras')
```
